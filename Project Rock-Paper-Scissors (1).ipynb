```python
import tkinter as tk
import random


user_choice = ""
comp_choice = ""
result_text = ""

def play():
    global user_choice, comp_choice, result_text
    user_choice = user_choice_entry.get()
    
    if user_choice not in choices:
        result_text = "Choix invalide."
    else:
        comp_choice = random.choice(choices)
        if user_choice == comp_choice:
            result_text = "C'est une égalité !" 
        elif (
            (user_choice == "Pierre" and comp_choice == "Ciseaux") or
            (user_choice == "Papier" and comp_choice == "Pierre") or
            (user_choice == "Ciseaux" and comp_choice == "Papier")
        ):
            result_text = "Vous gagnez !"
        else:
            result_text = "L'ordinateur gagne !"

    result_label.config(text=result_text)

def reset():
    global user_choice, comp_choice, result_text
    user_choice = ""
    comp_choice = ""
    result_text = ""
    user_choice_entry.delete(0, tk.END)
    result_label.config(text="")

def exit_game():
    window.destroy()

window = tk.Tk()
window.title("Jeu Pierre-Papier-Ciseaux")
window.geometry("400x400")
window.configure(bg="blue")

title_label = tk.Label(window, text="Jeu Pierre-Papier-Ciseaux", font=("Arial", 16))
title_label.pack(pady=20)

prompt_label = tk.Label(window, text="Choisissez : Pierre, Papier ou Ciseaux", font=("Arial", 12))
prompt_label.pack()

user_choice_entry = tk.Entry(window, font=("Arial", 12))
user_choice_entry.pack()

play_button = tk.Button(window, text="Jouer", command=play, font=("Arial", 12))
play_button.pack()

result_label = tk.Label(window, text="", font=("Arial", 12))
result_label.pack()

reset_button = tk.Button(window, text="Réinitialiser", command=reset, font=("Arial", 12))
reset_button.pack()

exit_button = tk.Button(window, text="Quitter", command=exit_game, font=("Arial", 12))
exit_button.pack()

choices = ["Pierre", "Papier", "Ciseaux"]
comp_pick = random.choice(choices)

window.mainloop()
```
